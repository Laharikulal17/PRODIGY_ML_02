```python
import os
import zipfile
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
```

```python
dataset_path = '/content/drive/MyDrive/archive (2)'
```

```python
datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)
```

```python
image_size = (128, 128)
batch_size = 32

image_size = (128, 128)
batch_size = 32

train_generator = datagen.flow_from_directory(
    os.path.join(dataset_path),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
)
test_generator = datagen.flow_from_directory(
    os.path.join(dataset_path),
    target_size=image_size,
    batch_size=batch_size,
    class_mode='categorical',
)
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, preprocessing

# ... (Your existing code for datagen, train_generator, test_generator)

# Define your CNN model
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 3))) # Match input shape with image_size
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())

# Calculate the correct number of neurons for the Dense layer
# You'll need to determine the output shape after the Flatten layer.
# A common way is to add a print statement after the Flatten layer:
model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu')) # Adjust this number based on the output of the print statement
# Add an output layer (adjust num_classes to match your dataset)
num_classes = 2  # Replace with the actual number of classes in your dataset
model.add(layers.Dense(num_classes, activation='softmax'))

# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print the model summary to verify shapes
model.summary()

# Train the model (adjust steps_per_epoch based on your dataset size)
model.fit(train_generator, epochs=5, validation_data=test_generator, steps_per_epoch=len(train_generator))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np

Y_pred = model.predict(test_generator)
y_pred = np.argmax(Y_pred, axis=1)

y_true = test_generator.classes

conf_matrix = confusion_matrix(y_true, y_pred)
print('Confusion Matrix')
print(conf_matrix)

class_report = classification_report(y_true, y_pred, labels=np.unique(y_true))
print('Classification Report')
print(class_report)
```
